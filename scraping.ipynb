```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
A,B,C,D,E,F,G,H,I,J,K,L=([] for i in range(12))
for i in range(100):
    url = "http://insidertrading.org/index.php?sort_by=acceptance_datetime&asc=&symbol=&date_from=2016-08-18&date_to=2017-12-26&submit=+GO+&page=" + str(i+1)
    page = requests.get(url)

    # Get the page
    doc = BeautifulSoup(page.text, 'html.parser')
    


    for row in doc.find_all("tr",class_='odd'):
    
        cells = row.find_all('td')
   
     
        if len(cells)==12: #Only extract table body not heading
            A.append(cells[0].find(text=True))
            B.append(cells[1].find(text=True))
            C.append(cells[2].find(text=True))
            D.append(cells[3].find(text=True))
            E.append(cells[4].find(text=True))
            F.append(cells[5].find(text=True))
            G.append(cells[6].find(text=True))
            H.append(cells[7].find(text=True))
            I.append(cells[8].find(text=True))
            J.append(cells[9].find(text=True))
            K.append(cells[10].find(text=True))
            L.append(cells[11].find(text=True))
            
    for row in doc.find_all("tr",class_='even'):
    
        cells = row.find_all('td')
   
     
        if len(cells)==12: #Only extract table body not heading
            A.append(cells[0].find(text=True))
            B.append(cells[1].find(text=True))
            C.append(cells[2].find(text=True))
            D.append(cells[3].find(text=True))
            E.append(cells[4].find(text=True))
            F.append(cells[5].find(text=True))
            G.append(cells[6].find(text=True))
            H.append(cells[7].find(text=True))
            I.append(cells[8].find(text=True))
            J.append(cells[9].find(text=True))
            K.append(cells[10].find(text=True))
            L.append(cells[11].find(text=True))
            
    if (i%100==0): # To check whether my code is working or not
        print (len(A))
            
```

```python
df=pd.DataFrame(A,columns=['Buy/Sell'])
df['Transaction Date']=B
df['Acceptance Date Time']=C
df['Issuer Name']=D
df['Issuer Trading Symbol']=E
df['Reporting Owner Name']=F
df['Reporting Owner Relationship']=G
df['Transcation Share']=H
df['Price Per share']=I
df['Total Value']=J
df['Shares Owned Following Transaction']=K
df['Form']=L
```

```python
healthcare=pd.read_csv("/data/Industry/Healthcare.csv")
basic_industries=pd.read_csv("/data/Industry/basic industries.csv")
capital_goods=pd.read_csv("/data/Industry/capital goods.csv")
consumer_durables=pd.read_csv("/data/Industry/Consumer Durables.csv")
technology=pd.read_csv("/data/Industry/technology.csv")
energy=pd.read_csv("/data/Industry/Energy.csv")
consumer_services=pd.read_csv("/data/Industry/Consumer services.csv")
finance=pd.read_csv("/data/Industry/finance.csv")
public_utilities=pd.read_csv("/data/Industry/Public utilities.csv")
transportation=pd.read_csv("/data/Industry/Transportation.csv")
miscellaneous=pd.read_csv("/data/Industry/Miscellaneous.csv")
```

```python
sector=pd.concat([healthcare,basic_industries,capital_goods,consumer_durables,technology,energy,consumer_services,finance,public_utilities,transportation,miscellaneous])
sector=sector.reset_index()
```

```python
industry=[]
market_cap=[]
for x in range(len(df['Issuer Trading Symbol'])):
    l=list(sector['Symbol'])
    if (df['Issuer Trading Symbol'][x] in l):
        a=l.index(df['Issuer Trading Symbol'][x])
        industry.append(sector['Sector'][a])
        market_cap.append(sector['MarketCap'][a])
        
    
    else:
        industry.append('unknown')
        market_cap.append('unknown')
df['sector']=industry
df['market_cap']=market_cap
    
```

```python
df.to_csv('/data/insiderdata.csv')
```
